```python
import pandas as pd
from itertools import takewhile
from tqdm import tqdm
```

```python
filename = 'asciimesh/d2022.bin'

def find_bias_terms(lst):
    synonms = []
    for i in lst:
        if ('ENTRY =' in i) & ('Bias' in i): # the ENTRY/PRINT ENTRY is always uppercase
            synonms.append(i.split(' = ')[1].split('|')[0])
    if len(synonms) > 0:
        df = pd.DataFrame.from_records([j.split(' = ') for j in lst]).set_index(0).T
        df= df[['MH','UI','MS']]
        df.columns = ['MeSH term', 'Unique identifier', 'Description']
        df['Synonyms'] = '; '.join(synonms)
        return(df)
            
with open(filename) as file:
    lines = file.readlines()
    par_lst = []
    lst = []
    df_lst = []
    for line in lines:
        if line.startswith('*NEWRECORD'):
            par_lst.append(lst)
            df_lst.append(find_bias_terms(lst))
            lst = []
        else:
            lst.append(line.rstrip())
df = pd.concat(df_lst).reset_index(drop=True)
df.to_csv('MeSH_bias_table.tsv',sep='\t',index=None)
```
